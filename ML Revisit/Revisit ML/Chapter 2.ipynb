```python
iris = pd.read_csv('../assets/ML/iris.csv',header=None,encoding='utf-8')

def tweak_iris(df_:pd.DataFrame)->pd.DataFrame:
    return(
        df_.rename(columns={
            0:'sepal_length',
            1:'sepal_width',
            2:'petal_length',
            3:'petal_width',
            4:'species'
        }).astype({
            'sepal_length':'float64',
            'sepal_width':'float64',
            'petal_length':'float64',
            'petal_width':'float64',
            'species':'category'
        })
    )

df = tweak_iris(iris)
df.info()
```

```python
fig,ax = plt.subplots(1,1,figsize=(12,5))
y = df.loc[:100,'species'].values
y = np.where(y=='Iris-setosa',0,1)

X = df.loc[:100,['sepal_length','petal_length']].values

ax.scatter(X[:50,0],X[:50,1],color='red',marker='o',label='Setosa')
ax.scatter(X[50:100,0],X[50:100,1],color='blue',marker='s',label='Versicolor')
ax.set_xlabel('Sepal length [cm]')
ax.set_ylabel('Petal length [cm]')
ax.legend(loc='best')
fig.show()
```

```python
ppn = Perceptron(lr=0.1,n_iter=10,random_seed=5)
ppn.fit(X=X,y=y)

print(f"w1&w2 are{ppn.w_}, b are {ppn.b_}")

fig,ax = plt.subplots(1,1,figsize=(7,3))
ax.plot(range(1,len(ppn.errors_)+1), ppn.errors_, marker='o')
plt.xlabel('Epochs')
plt.ylabel('Number of updates')
fig.show()
```

```python
plot_decision_regions(X,y,classifer=ppn)
plt.xlabel('Sepal legth [cm]')
plt.ylabel('Petal legth [cm]')
plt.legend(loc='upper left')
plt.show()
```

**Mini-batch gradient descent**

- A compromise between full batch GD and SGD is so-called mini-batch gradient descent. 
- Mini-batch gradient descent can be understood as applying full batch GD to smaller subsets of the training data, 
    - i.e) The advantage over full batch GD is that convergence is reached faster  via mini-batches because of the more frequent weight updates
